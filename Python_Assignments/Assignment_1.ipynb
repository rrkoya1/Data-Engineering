1) Reverse a list (no .reverse() and no slicing [::-1])

```python
# FOR loop
fruits = ['banana', 'orange', 'mango', 'lemon']
rev_for = []
for i in range(len(fruits) - 1, -1, -1):
    rev_for.append(fruits[i])
print("1) Reverse list (FOR):", rev_for)
```

```python
# While-loop (in-place swap using two pointers)
fruits2 = ['banana', 'orange', 'mango', 'lemon']  # fresh copy
left, right = 0, len(fruits2) - 1
while left < right:
    fruits2[left], fruits2[right] = fruits2[right], fruits2[left]
    left += 1
    right -= 1
print("1) Reverse list (WHILE):", fruits2)
```

 2) FizzBuzz (1 to n)

```python
n = int(input("\n2) Enter n for FizzBuzz: "))

# For-loop
fb_for = []
for i in range(1, n + 1):
    if i % 15 == 0:
        fb_for.append("FizzBuzz")
    elif i % 3 == 0:
        fb_for.append("Fizz")
    elif i % 5 == 0:
        fb_for.append("Buzz")
    else:
        fb_for.append(str(i))
print("2) FizzBuzz (FOR):", ", ".join(fb_for))
```

```python
# While-loop
fb_while = []
i = 1
while i <= n:
    if i % 15 == 0:
        fb_while.append("FizzBuzz")
    elif i % 3 == 0:
        fb_while.append("Fizz")
    elif i % 5 == 0:
        fb_while.append("Buzz")
    else:
        fb_while.append(str(i))
    i += 1
print("2) FizzBuzz (WHILE):", ", ".join(fb_while))
```

3) Factorial of n

```python
n = int(input("\n3) Enter n for factorial (>=0): "))
if n < 0:
    print("3) Factorial: n must be >= 0")
else:
    # For-loop
    fact_for = 1
    for x in range(2, n + 1):
        fact_for *= x
    print("3) Factorial (FOR):", fact_for)

    # While-loop
    fact_while = 1
    temp = n
    while temp > 1:
        fact_while *= temp
        temp -= 1
    print("3) Factorial (WHILE):", fact_while)
```

4) snake_case → camelCase

```python
s = input("\n4) Enter snake_case string: ")
result = []
cap_next = False

for ch in s:
    if ch == "_":
        cap_next = True
    else:
        if cap_next:
            result.append(ch.upper())
            cap_next = False
        else:
            result.append(ch)

print("4) camelCase:", "".join(result))
```

5) Floyd’s Triangle (rows)

```python
rows = int(input("\n5) Enter number of rows for Floyd’s Triangle: "))
num = 1
print("5) Floyd’s Triangle:")
for r in range(1, rows + 1):
    line = []
    for _ in range(r):
        line.append(str(num))
        num += 1
    print(" ".join(line))
```

6) Sum of digits (using % and //)

```python
x = int(input("\n6) Enter an integer to sum its digits: "))
x = abs(x)
total = 0
while x > 0:
    total += x % 10
    x //= 10
print("6) Sum of digits:", total)
```

 7) Until user enters a negative (count, sum, average)

```python
print("\n7) Enter numbers (negative to stop):")
count = 0
total = 0

while True:
    val = int(input("   Number: "))
    if val < 0:
        break
    total += val
    count += 1

avg = total / count if count > 0 else 0
print(f"7) Count={count}, Sum={total}, Average={avg}")
```

8) Count vowels in a string (total + per-vowel) using while loop

```python
text = input("Enter a string to count vowels: ")

vowels = "aeiou"
counts = {v: 0 for v in vowels}
total_vowels = 0

i = 0
text_lower = text.lower()

while i < len(text_lower):
    ch = text_lower[i]
    if ch in counts:
        counts[ch] += 1
        total_vowels += 1
    i += 1

print("Total vowels:", total_vowels)
print("Per vowel:", counts)
```

9) Second largest DISTINCT number (no sorting)

```python
nums = [10, 5, 20, 8, 20, 3]  
largest = None
second = None

for val in nums:
    if largest is None or val > largest:
        if largest != val:
            second = largest
        largest = val
    elif val != largest and (second is None or val > second):
        second = val

print("\n9) List:", nums)
print("9) Second largest distinct:", second)
```

10) Two-sum (find indices of 2 numbers that add to target)

```python
nums = [2, 7, 11, 15]  
target = int(input("\n10) Enter target for Two-Sum: "))

seen = {}  # value -> index
answer = None

for i, val in enumerate(nums):
    need = target - val
    if need in seen:
        answer = (seen[need], i)
        break
    seen[val] = i

print("10) List:", nums)
print("10) Two-sum result (indices):", answer)
```
