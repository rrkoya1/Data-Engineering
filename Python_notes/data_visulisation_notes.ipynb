## 2) Create Sample Data (NumPy + Pandas)

We will create a simple student performance dataset for practice.


```python
np.random.seed(42)

df = pd.DataFrame({
    "Student": ["A", "B", "C", "D", "E", "F", "G", "H"],
    "Hours_Studied": [2, 3, 5, 1, 4, 6, 3, 7],
    "Marks": [55, 65, 82, 48, 74, 88, 69, 93],
    "Attendance": [78, 85, 95, 70, 88, 96, 84, 98],
    "Category": ["Avg", "Avg", "Top", "Low", "Avg", "Top", "Avg", "Top"]
})

df
```

## 5) Combining and Reshaping Data

### `concat`, `merge`, `join`, `transpose`


```python
df1 = pd.DataFrame({"ID": [1, 2], "Name": ["Alice", "Bob"]})
df2 = pd.DataFrame({"ID": [3, 4], "Name": ["Charlie", "Diana"]})

pd.concat([df1, df2], axis=0, ignore_index=True)
```

## 7) Different Types of Plots (Core Session Content)


### 7.1 Line Plot (Trend / Curves)

Use for trends over time or ordered data.


```python
days = [1, 2, 3, 4, 5]
temperature = [30, 32, 31, 35, 36]

plt.plot(days, temperature, marker="o")
plt.title("Temperature Trend")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.show()
```

### 7.2 Histogram

A histogram shows the distribution of numeric values using bins.


```python
marks = [55, 60, 61, 68, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 95]

plt.hist(marks, bins=5)
plt.title("Marks Distribution (Histogram)")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

### 7.3 Scatter Plot

Best for studying relationship/correlation between two numeric variables.


```python
hours = [1, 2, 3, 4, 5, 6, 7]
scores_vals = [50, 55, 65, 70, 78, 85, 90]

plt.scatter(hours, scores_vals)
plt.title("Hours Studied vs Score")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.show()
```

```python
corr_df = pd.DataFrame({"Hours": hours, "Scores": scores_vals})
corr_df.corr()
```

### 7.4 Bar Chart

Best for comparing categories.


```python
categories = ["Books", "Electronics", "Clothes"]
sales_values = [200, 350, 150]

plt.bar(categories, sales_values)
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

### 7.5 Pie Chart

Best for part-to-whole comparisons when there are only a few categories.


```python
sizes = [40, 35, 25]
labels = ["Books", "Electronics", "Clothes"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Category Share")
plt.show()
```

### 7.6 Box Plot

Shows median, quartiles, spread, and outliers.


```python
data = [10, 12, 12, 13, 14, 15, 16, 18, 50]  # 50 is an outlier

plt.boxplot(data)
plt.title("Box Plot Example")
plt.show()
```

### 7.7 Area Chart

Useful for cumulative trends and emphasizing magnitude.


```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 2, 6, 4]

plt.fill_between(x, y, alpha=0.4)
plt.plot(x, y)
plt.title("Area Chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

### 7.8 Subplots

Use subplots to compare multiple plots in the same figure.


```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 8, 27, 64]

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title("Square")

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("Cube")

plt.tight_layout()
plt.show()
```

### 7.9 Stacked Bar Chart

Shows category totals and composition together.


```python
stack_df = pd.DataFrame({
    "Class": ["A", "B", "C"],
    "Pass": [30, 25, 28],
    "Fail": [5, 8, 4]
})

stack_df.set_index("Class")[["Pass", "Fail"]].plot(kind="bar", stacked=True)
plt.title("Stacked Bar Chart")
plt.ylabel("Students")
plt.show()
```

### 7.10 Bubble Chart

A bubble chart is a scatter plot where bubble size represents a third variable.


```python
x = [1, 2, 3, 4]
y = [10, 15, 13, 17]
sizes = [100, 300, 200, 500]

plt.scatter(x, y, s=sizes, alpha=0.5)
plt.title("Bubble Chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## 8) Pandas Built-in Plotting

Pandas plotting uses Matplotlib internally and is very convenient for quick analysis.


```python
df_plot = pd.DataFrame({
    "Day": [1, 2, 3, 4, 5],
    "Sales": [100, 120, 90, 150, 170]
})

df_plot.plot(x="Day", y="Sales", kind="line", title="Sales Trend")
plt.show()
```

```python
df_plot.plot(x="Day", y="Sales", kind="bar", title="Sales by Day")
plt.show()
```

```python
df["Marks"].plot(kind="hist", bins=5, title="Marks Histogram")
plt.show()
```

```python
df[["Marks", "Hours_Studied", "Attendance"]].plot(kind="box")
plt.title("Box Plot (Multiple Columns)")
plt.show()
```

```python
df.plot(kind="scatter", x="Hours_Studied", y="Marks", title="Hours vs Marks")
plt.show()
```

## 9) Variance and Correlation

### Variance
Measures how spread out data points are from the mean.

### Correlation
Measures linear relationship between variables.


```python
print("Variance of Marks:", df["Marks"].var())
print("\nCorrelation Matrix:")
df[["Marks", "Hours_Studied", "Attendance"]].corr()
```
